## Simple Manual model
 By the above representation in graph we can able to predict the type of flower by its one of the feature i.e petal_length

```python
iris.target_names
```

```python
def single_feature_prediction(petal_length):
    #Predicts the Iris species given the petal length
    if petal_length < 2.5:
        return 0
    elif petal_length < 4.8:
        return 1
    else:
        return 2
```

```python
X_train[:, 2]
```

```python
manual_y_predictions = np.array([single_feature_prediction(val) for val in X_train[:, 2]])
```

```python
manual_model_accuracy = np.mean(manual_y_predictions == y_train) # it telling the accuracy of manual_y_predictions 
```

```python
print("Manual model accuracy: ", manual_model_accuracy)
```
