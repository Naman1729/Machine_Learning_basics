```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('canada_per_capita_income.csv')
df
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('money(US$)')
plt.scatter(df.year, df.money, color='red', marker='+')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['year']], df.money)
```

```python
reg.predict([[1990]])
```
